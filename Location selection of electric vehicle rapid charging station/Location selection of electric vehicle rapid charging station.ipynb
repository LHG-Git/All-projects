### 전기차대수

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans

# 전기차 대수
# 두 가지 feature를 대상
data_대수_급속 = datas[['전기차 대수', '급속충전기']]

# 정규화 진행
scaler = MinMaxScaler()
data_scale = scaler.fit_transform(data_대수_급속)
```

```python
from yellowbrick.cluster import KElbowVisualizer

model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,10))
visualizer.fit(data_scale)
```

```python
from sklearn.cluster import KMeans

k = 3

# 그룹 수, random_state 설정
model = KMeans(n_clusters = k, random_state = 10)

# 정규화된 데이터에 학습
model.fit(data_scale)

# 클러스터링 결과 각 데이터가 몇 번째 그룹에 속하는지 저장
data_대수_급속['cluster'] = model.fit_predict(data_scale)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (8, 8))

for i in range(k):
    plt.scatter(data_대수_급속.loc[data_대수_급속['cluster'] == i, '급속충전기'], data_대수_급속.loc[data_대수_급속['cluster'] == i, '전기차 대수'], 
                label = 'cluster ' + str(i))

plt.legend()
plt.title('K = %d results'%k , size = 15)
plt.xlabel('급속충전기 개수', size = 12)
plt.ylabel('전기차 대수', size = 12)
plt.show()
```
